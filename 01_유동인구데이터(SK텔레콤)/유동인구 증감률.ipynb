```python
def get_float_rate(x):
    if x.Total_y is not None:
        return (x.Total_x-x.Total_y)/x.Total_y
    else:
        return (x.Total_x-x.Total)/x.Total
```

```python
floating_pop_2020_rating_v2.fillna('널', inplace = True)
```

```python
floating_pop_2020_rating_v2['증감률']=0
for i in range(0, len(floating_pop_2020_rating_v2)):
    if floating_pop_2020_rating_v2.loc[i,'Total_y'] == '널':
        floating_pop_2020_rating_v2.loc[i,'증감률'] =(floating_pop_2020_rating_v2.loc[i,'Total_x']-floating_pop_2020_rating_v2.loc[i,'Total'])/floating_pop_2020_rating_v2.loc[i,'Total']
    else:
        floating_pop_2020_rating_v2.loc[i,'증감률'] =(floating_pop_2020_rating_v2.loc[i,'Total_x']-floating_pop_2020_rating_v2.loc[i,'Total_y'])/floating_pop_2020_rating_v2.loc[i,'Total_y']
    
```
