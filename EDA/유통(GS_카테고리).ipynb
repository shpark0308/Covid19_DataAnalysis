```python
parsed_ctg_GS = ctg_GS.groupby(["시","구","분석용상품대분류명"])[ctg_GS.columns[4:]].mean()
```

```python
parsed_ctg_GS
```

```python
t_parsed_ctg_GS = parsed_ctg_GS.T
```

```python
t_parsed_ctg_GS
```

```python
ctg_GS_2019 =  t_parsed_ctg_GS.iloc[:120,:32]
```

```python
ctg_GS_2020 =  t_parsed_ctg_GS.iloc[120:,:32]
```

```python
data_2019_02 = ctg_GS_2019[:28].mean()
data_2020_02 = ctg_GS_2020[:28].mean()
data_2019_03 = ctg_GS_2019[28:59].mean()
data_2020_03 = ctg_GS_2020[28:59].mean()
data_2019_04 = ctg_GS_2019[59:89].mean()
data_2020_04 = ctg_GS_2020[59:89].mean()
data_2019_05 = ctg_GS_2019[89:120].mean()
data_2020_05 = ctg_GS_2020[89:120].mean()
```

```python
data_2019_02
```

```python
data_2019_02.unstack()
data_2020_02.unstack()
data_2019_03.unstack()
data_2020_03.unstack()
data_2019_04.unstack()
data_2020_04.unstack()
data_2019_05.unstack()
data_2020_05.unstack()
```

```python
temp_data = pd.concat([data_2019_02, data_2019_03, data_2019_04, data_2019_05, data_2020_02, data_2020_03, data_2020_04, data_2020_05],axis=1)
temp_data
```

```python
temp_data.columns = ["201902","201903","201904","201905","202002","202003","202004","202005"]
```

```python
temp_data
```

```python
ctg_GS_2019 = temp_data.iloc[:32,:4]
```

```python
ctg_GS_2020 = temp_data.iloc[:32,4:]
```

```python
t_ctg_GS_2019 = ctg_GS_2019.T
t_ctg_GS_2019
```

```python
t_ctg_GS_2020 = ctg_GS_2020.T
t_ctg_GS_2020
```

```python
title =["간식", "마실거리", "사회활동", "식사", "임신/육아", "취미&여가활동", "헬스&뷰티", "홈&리빙"]
```

```python
X = np.arange(4)
fig = plt.figure(figsize = (18,40))
for i in range(0,8):
    plt.subplot(8,4,i+1)
    plt.bar(X+2.00, list(t_ctg_GS_2019[t_ctg_GS_2019.columns[i]]), label='2019',color = 'b', width = 0.25)
    plt.bar(X+2.25, list(t_ctg_GS_2020[t_ctg_GS_2020.columns[i]]), label='2020',color = 'r', width = 0.25)
    plt.legend(loc="best",frameon=False)
    plt.ylim(0,45)
    plt.title("대구_수성구_"+title[i])
    plt.xlabel("월")
    plt.ylabel("%(퍼센트)")
plt.show()
```
